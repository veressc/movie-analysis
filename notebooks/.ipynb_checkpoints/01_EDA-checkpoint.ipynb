```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv('../data/movies.csv', quotechar='"')

# Просмотр первых строк таблицы
print(df.head())

# Информация о датасете
print(df.info())

# Средний рейтинг
average_rating = df['rating'].mean()
print(f"Средний рейтинг фильмов: {average_rating:.2f}")

# Топ-10 фильмов
top_10_movies = df.sort_values(by='rating', ascending=False).head(10)
print("Топ-10 фильмов по рейтингу:")
print(top_10_movies[['title', 'rating']])

# График год-рейтинг
ratings_by_year = df.groupby('year')['rating'].mean()

plt.figure(figsize=(10, 5))
plt.plot(ratings_by_year.index, ratings_by_year.values, marker='o')
plt.title('Средний рейтинг фильмов по годам')
plt.xlabel('Год выпуска')
plt.ylabel('Средний рейтинг')
plt.grid(True)
plt.show()
```
